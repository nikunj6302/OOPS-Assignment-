```python
#Ans. 2
import abc

class Shape():
  @abc.abstractmethod
  def area(self):
    print("area of shape")

class Rectangle(Shape):
  def __init__(self,l,w):
    self.length = l
    self.width = w

  def area(self):
    return self.length * self.width

class Circle(Shape):
  def __init__(self,r):
    self.radius = r

  def area(self):
    return 3.14 * self.radius * self.radius

c = Circle(5)
print(c.area())
c = Shape()
c.area
```

```python
#Ans. 5
class BankAccount():
  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    self.__balance += amount
    print(self.__balance)

  def withdraw(self, amount):
    if amount <= self.__balance:
      self.__balance -= amount
      print(self.__balance)
    else:
      print("Insufficient balance")


jay = BankAccount(1000)
jay.deposit(500)
jay.withdraw(2000)
```

```python
#Ans. 6
import time

start_time = time.time()

class Instrument():
  def play(self):
    print("Instrument can play")


class Piano(Instrument):
  def play(self):
    print("Piano can play")

class Guitar(Instrument):
  def play(self):
    print("Guitar can play")

p = Piano()
p.play()
g = Guitar()
g.play()

result = 1
for i in range(1, 1000001):
    result += i

end_time = time.time()

run_time = end_time - start_time
print(f"The runtime of the code is: {run_time} seconds")
```

```python
#Ans. 10
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

a = Vector(5,6)
b = Vector(10,20)
c = a + b
print(c.x, c.y)
```

```python
#Ans. 12
class Student():
  def __init__(self, name, grade1,grade2,grade3,grade4,grade5):
    self.name = name
    self.maths = grade1
    self.science = grade2
    self.english = grade3
    self.history = grade4
    self.geography = grade5


  def average_grades(self):
    total = self.maths + self.science + self.english + self.history + self.geography
    avg = total/5
    print(avg)

a = Student("nik",95,70,55,45,60)
a.average_grades()
```

```python
#Ans.14
class Employee():
  def __init__(self,name,hours):
    self.name = name
    self.hours = hours

  def calcuate_slary(self):
    print("the salary is", 100*self.hours)

class Manager(Employee):
  def __init__(self,name,hours):
    super().__init__(name,hours)

  def calculate_slary(self):
    print("the salary is", 150*self.hours + 10,000)

nik = Employee("nik",10)
nik.calcuate_slary()
ajay = Manager("ajay",10)
ajay.calculate_slary()
```

```python
#Ans. 16
import abc

class Animal():
  @abc.abstractmethod
  def sound(self):
    print("Animal can make sound")

class Cow(Animal):
  def sound(self):
    print("Moo!")

class Sheep(Animal):
  def sound(self):
    print("Mehhh")

c = Cow()
c.sound()
s = Sheep()
s.sound()
```

```python
#Ans. 18
class House():
  def __init__(self,address,price):
    self.address = address
    self.price = price

  def house_info(self):
    print(f"Address: {self.address}, Price: {self.price}")

class Mension(House):
  def __init__(self,address,price,rooms):
    super().__init__(address,price)
    self.rooms = rooms

  def mension_info(self):
    print(f"Address: {self.address}, Price: {self.price}, Rooms: {self.rooms}")


a = House("23A, nayak nagar, gurukul, ahmedabad, gujrat", 4000000)
a.house_info()

b = Mension("179, jasmin app., usmanpura, ahmedabad, gujrat", 25000000,5)
b.mension_info()
```
